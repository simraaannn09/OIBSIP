```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Advertising.csv', encoding='latin')
```

```python
df.head(5)
```

```python
df.drop(columns=['Unnamed: 0' ],inplace= True)
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.pairplot(df,x_vars=['TV' , 'Newspaper', 'Radio'],y_vars='Sales',kind='scatter')
plot.show()
```

```python
df['Radio'].plot.hist(bins=10,color="green",xlabel="Radio")
```

```python
sns.heatmap(df.corr(),annot=True)
plot.show()
```

```python
#Training the model
Y_train, Y_test, X_train,X_test= train_test_split(df[['TV']], df[['Sales']],test_size=0.3,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
model=LinearRegression()
model.fit(X_train, Y_train)
```

```python
res=model.predict(X_test)
print(res)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.coef_
```

```python
model.intercept_
```
